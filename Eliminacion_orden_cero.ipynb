# Imagen I - sin zoom

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Cargar la imagen
imagen_i = cv2.imread('/content/test_holograma_/Pruebas_muestra_sangre/I_sangre.bmp', 0)  # Carga la imagen en escala de grises

# Aplicar la Transformada de Fourier
transformada_fourier = np.fft.fft2(imagen_i)
transformada_fourier_shifted = np.fft.fftshift(transformada_fourier)

# Calcular el espectro de amplitud
espectro_amplitud = np.abs(transformada_fourier_shifted)

# Mostrar la imagen original y el espectro de amplitud
plt.subplot(121), plt.imshow(imagen_i, cmap='gray')
plt.title('Imagen Original'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(np.log(1 + espectro_amplitud), cmap='gray')
plt.title('Espectro de Amplitud'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
# Crear un histograma
plt.hist(imagen_i, bins=5, edgecolor='black', alpha=0.7)
# "datos" es la lista de datos que se va a usar en el histograma.
# "bins" especifica el número de contenedores o intervalos en el histograma.
# "edgecolor" establece el color de los bordes de las barras del histograma.
# "alpha" controla la transparencia de las barras.

# Etiquetas y título
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.title('Histograma de Datos')
```

# Experimento sin zoom

```python
import numpy as np

# Carga las dos imágenes que deseas restar
I = cv2.imread('/content/test_holograma_/Pruebas_muestra_sangre/I_sangre.bmp')
O = cv2.imread('/content/test_holograma_/Pruebas_muestra_sangre/O_sangre.bmp')
R = cv2.imread('/content/test_holograma_/Pruebas_muestra_sangre/R_sangre.bmp')

# Convertir la imagen a escala de grises
gris_I = cv2.cvtColor(I, cv2.COLOR_BGR2GRAY)
gris_O = cv2.cvtColor(O, cv2.COLOR_BGR2GRAY)
gris_R  = cv2.cvtColor(R, cv2.COLOR_BGR2GRAY)

# Asegúrate de que ambas imágenes tengan las mismas dimensiones
if gris_I.shape == gris_O.shape == gris_R.shape:
    # Realiza la resta de imágenes
    resultado = cv2.add(gris_R**2, gris_O**2)
    resultados = cv2.subtract(gris_I, resultado)

    conjugada_R = np.conjugate(gris_R)[0,1]
    conjugada_O = np.conjugate(gris_O)[0,1]

    OR = np.dot(gris_O, conjugada_R)
    RO = np.dot(gris_R, conjugada_O)
    sum = cv2.add(OR, RO)


    holograma = 2*(sum)

    sum2 = cv2.add(gris_I, holograma)

    # Guarda el resultado en un archivo de imagen
    cv2.imwrite('resultados.jpg', resultado)
    cv2.imwrite('holograma.jpg', holograma)
    cv2.imwrite('I_GRIS.jpg', gris_I)
    cv2.imwrite('SUM2.jpg', sum2)

    # Muestra el resultado
    #cv2.imshow('Resultado', resultado)
    #cv2.waitKey(0)
    #cv2.destroyAllWindows()
else:
    print("Las imágenes tienen dimensiones diferentes y no se pueden restar.")
```

```python
# Crear un histograma
plt.hist(holograma, bins=5, edgecolor='black', alpha=0.7)
# "datos" es la lista de datos que se va a usar en el histograma.
# "bins" especifica el número de contenedores o intervalos en el histograma.
# "edgecolor" establece el color de los bordes de las barras del histograma.
# "alpha" controla la transparencia de las barras.

# Etiquetas y título
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.title('Histograma de Datos')
```
